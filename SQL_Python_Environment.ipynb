## More Complex SELECT Queries

```python
# Import data

!wget -O employees.sqlite3 "https://drive.google.com/uc?export=download&id=1ThJIPMRMlXOedWG2JIse0fod1wxc6XGu"
%sql sqlite:///employees.sqlite3
%config SqlMagic.style = '_DEPRECATED_DEFAULT'
```

```python
# Let's say we want to see common last names, e.g., frequency > 150 from the employees table
# We can use 'HAVING' with aggregate functions (after we use 'GROUP BY')

%%sql
SELECT last_name as 'Family Name'
FROM employees
GROUP BY last_name -- group by last name so we can use an aggregate function
HAVING COUNT(last_name) > 150;
```

```python
# Let's try to look at the first 20 salaries of employees that worked after august 20'th 1999 (not including)
# We can do this by using nested queries, i.e., an inner query which is nested within and outer query
# Note that our output still contains the slaries before the given date for employees that worked before and after the date

%%sql
SELECT emp_no as '# Employee',  salary as 'Salary', from_date
FROM salaries
WHERE emp_no IN (
  -- This is how the nested query looks
  SELECT emp_no
  FROM salaries
  WHERE from_date > 1999-08-20)
LIMIT 20;
```

```python
# We want to see all columns for the 5 most senior employees, i.e., by employment date
# We can again use a nested query for this

%%sql
SELECT *
FROM employees
WHERE emp_no IN (
  SELECT emp_no
  FROM salaries
  ORDER BY from_date
  LIMIT 5);
```

```python
# Now we want to see all emplyees, s.t. the 2'nd lettter of thier last name is an a
# We an use the 'LIKE' statement for this, which allows lookups by string queries

%%sql
SELECT *
FROM employees
WHERE last_name LIKE '_a%'
/* Note:The underscore '_' denotes that there is a letter before the 'a'
   The % denotes that there is some string (can also be empty) afetr the 'a'.
   This could just as well be replaced with an arbitrary number of %%% symbols */
;

```

## Joining tables - Inner Joins

```python
# Joining tables using Kartesian multiplication
# recall that Kartesian multiplication given us all possible cominations
# so we need to filter the rows to only those that the
# foreign key from the primary table is the same as primary key of the second table

%%sql

SELECT e.id, e.name as company_name, e.id_address, a.id_address, a.address
FROM entity e, address a
WHERE e.id_address = a.id_address
-- can denote entity by e and address by a, so we can reference them  faster when selecting columns
LIMIT 20;
```

```python
# Joining tables using 'JOIN ... ON' statements
# This is a different way of implemeting the same execution

%%sql

SELECT e.id, e.name as company_name, e.id_address, a.id_address, a.address
FROM entity e JOIN address a ON e.id_address = a.id_address
-- can again denote entity by e and address by a
LIMIT 20;
```

```python
# Now imagine a scenario when we want to join two tables, but we do not a foreign key
# We can however join them using a third "mediator" table that has foreign keys for both

%%sql

SELECT *
FROM intermediary i, entity e, assoc_inter_entity a
WHERE e.id = a.entity AND
      a.inter = i.id
      AND e.name = 'Big Data Crunchers Ltd.';

```

## **Joining tables - Outer Joins**

```python
# Say we want to keep all rows in one of our tables,
# even those with missing value in our second table.
# Then, we can use 'LEFT/RIGHT OUTER JOIN'
# Let's look at examples

# Left outer join - keeps all rows of the left table

%%sql

SELECT *
FROM entity LEFT OUTER JOIN address
ON entity.id_address = address.id_address;
```

```python
# Right outer join


%%sql


-- This is the general syntax, however it will not run since
-- RIGHT OUTER JOIN is not currently supported in this python environment

/*
SELECT *
FROM entity RIGHT OUTER JOIN address
ON entity.id_address = entity.id_address;
*/

-- Instead we can simply reverse the order of the tables and use LEFT OUTER JOIN


SELECT *
FROM address LEFT OUTER JOIN entity
ON entity.id_address = address.id_address;

```

## **Joining tables - Natural Joins**

```python
# We can also use natural joins, which can sometimes be the most intuitive


%%sql

SELECT *
FROM entity NATURAL JOIN address;

/*Note that we did not have to specify the column on which to join the tables.
Instead, the only column with the same name was selected*/
```

## **Group by**

```python
# In order to use 'GROUP BY' we need:
# a. a grouping variable
# b. a aggregating function (or a few)

## Note that without using an aggregating function, we get the same output as

# /*
# SELECT DISTINCT status
# FROM entity;
# */


%%sql

SELECT status
FROM entity
GROUP BY status;
```

```python
# Let's try only using an aggregating function
# This will give us the number of total rows in the table

%%sql

SELECT COUNT(*)
FROM entity;
```

```python
# Let's use both now
# We can use count and group by to obtain category frequencies
# We can also do this using multiple grouping variables,
# in order to obtain contingency tables

%%sql

SELECT status, jurisdiction, count(*) as frequency
FROM entity
GROUP BY status, jurisdiction;
```

```python
# The tables dept_emp and Dept_manager have the exact same variables
# We can take the union of them to combine them
# We can use the AS source statment to track which rows are employees and which are managers

%%sql

SELECT 'Employee' AS source, *
FROM dept_emp UNION
SELECT 'Manager' AS source, *
FROM Dept_manager;
```

```python
# Next, we want to look at the max salary of the employee born on January 31'st 1964

%%sql

SELECT e.first_name, e.last_name, MAX(s.salary)
FROM employees e, salaries s
WHERE e.emp_no = s.emp_no AND e.birth_date = '1964-01-31'
GROUP BY e.first_name, e.last_name;
```

```python
# Now let's look at the different departments that employee number 499902 worked at
# and let's also get the dates that they started and finished i every department

# The employee name is in employees
# The employee department is in dept_emp
# The department name is in departments

# Note that we could have also done this via:

# SELECT e.emp_no, e.first_name, e.last_name, de.from_date, de.to_date, de.dept_no, d.dept_name
# FROM employees e JOIN dept_emp de, departments d
# ON e.emp_no = de.emp_no AND de.dept_no = d.dept_no
# WHERE e.emp_no = 499902;

%%sql

SELECT e.emp_no, e.first_name, e.last_name, de.from_date, de.to_date, de.dept_no, d.dept_name
FROM employees e, dept_emp de, departments d
WHERE e.emp_no = de.emp_no AND de.dept_no = d.dept_no AND e.emp_no = 499902;
```

```python

# Let's now compare the mean salaries of employees by gender

%%sql

SELECT e.gender, ROUND(AVG(s.salary),2)
FROM employees e, salaries s
WHERE e.emp_no = s.emp_no
GROUP BY e.gender;
```

```python
# Let's say we want to count how many males and females have larger than average maximal salary

%%sql

DROP TABLE max_salaries;

CREATE TABLE max_salaries AS
SELECT e.emp_no, MAX(s.salary) max_salary
FROM employees e JOIN salaries s ON e.emp_no = s.emp_no
GROUP BY e.emp_no;

SELECT e.gender, COUNT(*)
FROM employees e, max_salaries ms
WHERE e.emp_no = ms.emp_no
AND ms.max_salary > (SELECT AVG(max_salary) FROM max_salaries)
GROUP BY e.gender;
```
